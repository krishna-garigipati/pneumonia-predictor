```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
from tensorflow.keras.models import load_model
```

```python
model = load_model('final_pneumonia_model.h5')
```

```python
def pre_process(image_path):
    img = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
    img = cv2.resize(img, (224,224))
    img = cv2.cvtColor(img, cv2.COLOR_GRAY2RGB)
    img = img/255.0
    img = np.expand_dims(img, axis=0)
    return img
```

```python
img_path = "image.png"

img = pre_process(img_path)
```

```python
img_display = cv2.imread(img_path , cv2.IMREAD_GRAYSCALE)
plt.figure(figsize=(5,5))

plt.imshow(img_display, cmap='gray')
plt.axis('off')
```

```python
prediction = model.predict(img)
```

```python
# Interpret the result and print accuracy
threshold = 0.5
if prediction[0][0] >= threshold:
    print(f" Pneumonia Detected!")
else:
    print(f" Normal Lung Detected ")
```
